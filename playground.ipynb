```python
from configparser import ConfigParser
from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt
from datetime import date

parser = ConfigParser()
_ = parser.read('nb.cfg')

api = SentinelAPI(parser.get('sentinel', 'username'), parser.get('sentinel','password'),'https://scihub.copernicus.eu/dhus')
footprint = geojson_to_wkt(read_geojson('map.geojson'))
```

```python

# WORKs but too much data
products = api.query(footprint,
                     platformname = 'Sentinel-2',
                    cloudcoverpercentage = (0, 10))

# THIS IS OFFLINE
#products = api.query(footprint,
#                     date = ('20151001', date(2016, 2, 1)),
#                     platformname = 'Sentinel-2',
#                     cloudcoverpercentage = (0, 10))
```

```python
# convert to Pandas DataFrame
products_df = api.to_dataframe(products)

products_df
```

```python
# sort and limit to first 5 sorted products
products_df_sorted = products_df.sort_values(['cloudcoverpercentage', 'ingestiondate'], ascending=[False, False])
products_df_sorted = products_df_sorted.head(2)

# download sorted and reduced products
api.download_all(products_df_sorted.index)
```
